# Redis - Gan Khai Li

```python
from pyspark.sql import SparkSession
import redis

#Initialize Spark session
spark = SparkSession.builder.appName("SaveToRedis").getOrCreate()
```

```python
singkatan_df = spark.read.csv("singkatan.csv", header=True, inferSchema=True)
singkatan_df.show()
```

```python
# Create a connection to the Redis server
r = redis.Redis(host='localhost', port=6379, db=0)
```

```python
# Collect the data and store it in Redis
for row in singkatan_df.collect():
    abbreviation = row['singkatan']
    word = row['perkataan']
    r.set(abbreviation, word)
```

```python
keys = r.keys()

for key in keys:
    try:
        abbreviation = key.decode('utf-8')  
        word = r.get(key).decode('utf-8')  
        print(f"Abbreviation: {abbreviation}, Word: {word}")
    except UnicodeDecodeError:
        print(f"Error decoding key or value: {key}")
```

```python
spark.stop()
```
